```python
import math
import json
import pandas
```

```python
def shannon_entropy(string):
    # get probability of chars in string
    prob = [ float(string.count(c)) / len(string) for c in dict.fromkeys(list(string)) ]

    # calculate the entropy
    entropy = - sum([ p * math.log(p) / math.log(2.0) for p in prob ])
    return entropy
```

```python
# Read in our data and return a dataframe

data_path = "data/http_loki.log"
http_transactions_raw = None
with open(data_path, 'r') as data_f:
    http_transactions_raw = data_f.readlines()
if not http_transactions_raw:
    raise Exception(f'File ({data_path}) is empty.')

http_transactions =  \
    pandas.DataFrame([json.loads(transaction) for transaction in http_transactions_raw if transaction.strip()])

# Convert mime-types to string
http_transactions.loc[:, 'host_string_entropy'] = http_transactions.host.map(lambda x: shannon_entropy(x) if isinstance(x, str) else 0)
http_transactions.sort_values(by='host_string_entropy', ascending=False)
http_transactions[['host','host_string_entropy']]
```
